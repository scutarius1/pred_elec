```python
df_temp=pd.read_csv("output2.csv" , sep = ',' )
df_temp
```

```python
df_temp = df_temp.drop(columns=['X', 'Y'])
df_temp = df_temp.drop(columns=['i', 'j'])
```

```python
data_summary = pd.DataFrame({
    'NaN Count': df_temp.isnull().sum(),
    'Data Type': df_temp.dtypes})
print(data_summary)
df_temp = df_temp.dropna()
df_temp
```

```python
df_temp["C°"] = df_temp["tasAdjust"] - 273.15
df_temp
```

```python
# Définition des zones géographiques sous forme de DataFrame
regions = [
    ("Auvergne-Rhône-Alpes", 44, 46.5, 2, 7),
    ("Bourgogne-Franche-Comté", 46.5, 48.5, 2, 6.5),  # Ajusté !
    ("Bretagne", 47, 49, -5.5, -1),
    ("Centre-Val de Loire", 46, 48, 0, 3),
    ("Grand Est", 48, 50, 4, 8),
    ("Hauts-de-France", 49, 51, 1, 4),
    ("Normandie", 48, 50, -1.8, 1.7),
    ("Nouvelle-Aquitaine", 43, 47, -1.8, 1.5),
    ("Occitanie", 42, 45, -1, 4),
    ("Pays de la Loire", 46, 48, -2, 1),
    ("Provence-Alpes-Côte d'Azur", 43, 45, 4, 7.5),
    ("Île-de-France", 48, 49, 2, 3),
]

df_regions = pd.DataFrame(regions, columns=["region", "lat_min", "lat_max", "lon_min", "lon_max"])
df_regions


```

```python
# Attribuer la région en filtrant les plages lat/lon
df_temp["region"] = None  # Nouvelle colonne vide

for _, row in df_regions.iterrows():
    mask = (
        (df_temp["lat"] >= row["lat_min"]) & 
        (df_temp["lat"] <= row["lat_max"]) & 
        (df_temp["lon"] >= row["lon_min"]) & 
        (df_temp["lon"] <= row["lon_max"])
    )
    df_temp.loc[mask, "region"] = row["region"]

# Calculer la moyenne par région et jour
df_mean_temp = df_temp.groupby(["time", "region"], as_index=False)["C°"].mean()

```

```python
# Calculer le nombre de lignes par région
nombre_lignes_par_region = df_temp["region"].value_counts()

# Afficher le nombre de lignes par région
print(nombre_lignes_par_region)
```

```python
df_idf_2024 = df_mean_temp[(df_mean_temp["region"] == "Bretagne") & (df_mean_temp["time"].str.startswith("2029"))]

moyenne_idf_2024 = df_idf_2024["C°"].mean()
print(f"Moyenne des températures en ....en 2024 : {moyenne_idf_2024:.2f}°C")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ajouter une colonne "year" pour extraire l'année depuis "time"
df_mean_temp["year"] = df_mean_temp["time"].str[:4].astype(int)

plt.figure(figsize=(14, 8))

# Tracé des températures moyennes pour chaque région
sns.lineplot(data=df_mean_temp, x="year", y="C°", hue="region", style="region", markers=True, dashes=False)

plt.xlabel("Année")
plt.ylabel("Température Moyenne (°C)")
plt.title("Évolution de la Température Moyenne par Année et Région")
plt.legend(title="Régions", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)

plt.show()
```

## CHECK COHERENCE DONNEES PASSEES ET FUTURES

```python
df=pd.read_csv("../DataSet/COMPILATION_CONSO_TEMP_POP_2.csv" , sep = ';' )
```

```python
df
```

```python
colonnes_a_supprimer = ['Date - Heure', 'TMin (°C)', 'TMax (°C)', 'POP (en milliers)']
df = df.drop(colonnes_a_supprimer, axis=1)
df['Date + Heure'] = pd.to_datetime(df['Date + Heure'])
df['Date'] = df['Date + Heure'].dt.date
df_temp_moy = df.groupby(['Région', 'Année', 'Date'])['TMoy (°C)'].mean().reset_index()
print(df_temp_moy)
```

```python
df_temp_moy.dtypes
# Convertir la colonne 'Date' et 'time' en type datetime
df_temp_moy['Date'] = pd.to_datetime(df_temp_moy['Date'])

df_mean_temp['time']= pd.to_datetime(df_mean_temp['time'])
```

```python
regions_selectionnees = ["Bretagne"]
#"Auvergne-Rhône-Alpes"
#"Bourgogne-Franche-Comté"
#Provence-Alpes-Côte d'Azur
# Filtrer df_temp_moy pour 2020 et les régions sélectionnées
df_2020 = df_temp_moy[
    (df_temp_moy["Année"] == 2024)
    & (df_temp_moy["Région"].isin(regions_selectionnees))
]

# Préparer les données pour le tracé
df_2020["month"] = df_2020["Date"].dt.month
df_2020["day"] = df_2020["Date"].dt.day
df_2020["date_index"] = pd.to_datetime(
    df_2020["month"].astype(str) + "/" + df_2020["day"].astype(str),
    format="%m/%d",
    errors="coerce",
)
df_2020 = df_2020.dropna(subset=["date_index"])
df_2020["region_annee"] = df_2020["Région"] + " 2020"

# Préparer les données de 2026
df_mean_temp["time"] = pd.to_datetime(df_mean_temp["time"], errors="coerce")
df_mean_temp["year"] = df_mean_temp["time"].dt.year
df_mean_temp["month"] = df_mean_temp["time"].dt.month
df_mean_temp["day"] = df_mean_temp["time"].dt.day
df_mean_temp["date_index"] = pd.to_datetime(
    df_mean_temp["month"].astype(str) + "/" + df_mean_temp["day"].astype(str),
    format="%m/%d",
    errors="coerce",
)
df_mean_temp = df_mean_temp.dropna(subset=["date_index"])
df_2026 = df_mean_temp[
    (df_mean_temp["year"] == 2024)
    & (df_mean_temp["region"].isin(regions_selectionnees))
]
df_2026["region_annee"] = df_2026["region"] + " 2026"

# Combiner les DataFrames et renommer les colonnes
df_2020_renamed = df_2020.rename(
    columns={"date_index": "date", "TMoy (°C)": "temperature", "Région": "region"}
)
df_2026_renamed = df_2026.rename(
    columns={"date_index": "date", "C°": "temperature"}
)

df_combined = pd.concat([df_2020_renamed, df_2026_renamed])

plt.figure(figsize=(14, 8))

# Tracé des températures moyennes
sns.lineplot(
    data=df_combined,
    x="date",  # Utiliser la date indexée comme axe des abscisses
    y="temperature",
    hue="region_annee",
    style="region_annee",
    markers=False,
    dashes=False,
)

plt.xlabel("Jour de l'année")
plt.ylabel("Température Moyenne (°C)")
plt.title("Évolution de la Température Moyenne (2020 et 2026) par Région")
plt.legend(title="Régions et Années", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.grid(True)
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter("%b"))  # Formatter l'axe des abscisses pour afficher uniquement les mois

plt.show()
```

```python
import pandas as pd

# Calcul des moyennes par année pour df_temp_moy
df_temp_moy["Année"] = df_temp_moy["Année"].astype(int)
moyennes_temp_moy = df_temp_moy.groupby("Année")["TMoy (°C)"].mean().reset_index()

# Calcul des moyennes par année pour df_mean_temp
df_mean_temp["time"] = pd.to_datetime(df_mean_temp["time"])
df_mean_temp["year"] = df_mean_temp["time"].dt.year
moyennes_mean_temp = df_mean_temp.groupby("year")["C°"].mean().reset_index()

# Fusion des deux tableaux
moyennes_combinées = pd.merge(
    moyennes_temp_moy, moyennes_mean_temp, left_on="Année", right_on="year", how="outer"
)

# Sélection des années de 2016 à 2036
moyennes_finales = moyennes_combinées[
    (moyennes_combinées["Année"].fillna(moyennes_combinées["year"]) >= 2016)
    & (moyennes_combinées["Année"].fillna(moyennes_combinées["year"]) <= 2036)
]

# Remplacer les NaN dans la colonne 'Année' par les valeurs de 'year'
moyennes_finales["Année"] = moyennes_finales["Année"].fillna(
    moyennes_finales["year"]
).astype(int)

# Suppression de la colonne 'year' en double
moyennes_finales = moyennes_finales.drop(columns=["year"])

# Renommage des colonnes
moyennes_finales = moyennes_finales.rename(
    columns={"Année": "Année", "TMoy (°C)": "Moyenne TMoy (°C)", "C°": "Moyenne tasAdjust"}
)

# Affichage du tableau
print(moyennes_finales)
```
