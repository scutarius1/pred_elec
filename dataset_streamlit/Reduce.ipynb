```python
import pandas as pd
df_cons = pd.read_csv("eco2mix-regional-cons-def.csv", sep=';',low_memory=False )
```

```python
colonnes_a_supprimer = ['Column 30','Stockage batterie', 'Déstockage batterie','Nature' ]
df_cons.drop(columns=colonnes_a_supprimer, inplace=True)
#df_cons['Echange Import (MW)'] = df_cons['Ech. physiques (MW)'].apply(lambda x: x if x > 0 else 0)
#df_cons['Echange Export (MW)'] = df_cons['Ech. physiques (MW)'].apply(lambda x: abs(x) if x < 0 else 0)
df_cons['Eolien (MW)'] = pd.to_numeric(df_cons['Eolien (MW)'], errors='coerce')
```

```python
# Tri par région (colonne 'Région')
df_sorted = df_cons.sort_values(by='Date - Heure')

# Réinitialiser les index après tri
df_sorted = df_sorted.reset_index(drop=True)

df_cleaned = df_sorted[df_sorted['Heure'].str.endswith(':00')]
df_cleaned['Consommation_MWh'] = df_cleaned['Consommation (MW)'] * 0.5
# Supprimer une ligne sur deux (conserver les lignes d'indice pair)
#df_cleaned = df_sorted.iloc[::2]
df_cleaned
```

```python
# Exporter en CSV
df_cleaned.to_csv("eco2mix-regional_reduced.csv", sep=',', index=False)
df_cleaned
```
